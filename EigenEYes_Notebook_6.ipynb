```python
import cv2
import numpy as np
import os
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Function to load images from a folder
def load_images(folder_path):
    images = []
    for filename in os.listdir(folder_path):
        img_path = os.path.join(folder_path, filename)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        images.append(img)
    return images

```

```python

def extract_green_channel(image):
    return image[:, :, 1]


def resize_image(image, size=(128, 128)):
    return cv2.resize(image, size)


def create_eigeneyes_hr(folder_path):

    images = load_images(folder_path)
    '''green channel extract and load resize'''
    green_channel_images = [extract_green_channel(img) for img in images]
    resized_images = [resize_image(img) for img in green_channel_images]


    data = np.array([img.flatten() for img in resized_images]) ; '''2D array'''

    '''principal component analysis step'''
    pca = PCA()
    pca.fit(data)

    '''mean eigenEYE for HR and top 10 eigenEYEs'''
    mean_face = pca.mean_.reshape(128, 128)
    top_eigeneyes = pca.components_[:10]

    plt.figure(figsize=(3, 3))

    '''mean display'''
    plt.imshow(mean_face, cmap='gray')
    plt.title('Mean Eye HR')
    plt.axis('off')

    '''top 10 features display'''
    plt.figure(figsize=(12, 8))

    for i, eigeneye in enumerate(top_eigeneyes):
        eigeneye_image = eigeneye.reshape(128, 128)
        plt.subplot(2, 5, i + 1)  # 3 rows, 4 columns
        plt.imshow(eigeneye_image, cmap='gray')
        plt.title(f'EigenEye_HR {i + 1}')

        plt.axis('off')

    plt.tight_layout()
    plt.show()


hreigen_path = "/content/drive/MyDrive/Eigen_Eyes/hr"

create_eigeneyes_hr(hreigen_path)
```
